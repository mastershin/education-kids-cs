```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
import requests

def get_image_from_url(url):
    """Get an image from a url and return it as a numpy array."""
    response = requests.get(url)
    # with open('image.jpg', 'wb') as f:
        # f.write(response.content)
    # Decode the image data
    img_array = np.array(bytearray(response.content), dtype=np.uint8)
    img_bgr = cv2.imdecode(img_array, cv2.IMREAD_COLOR)
    img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)
    return img_rgb

def read_image_from_file(path):
    # Read the image from the file
    img_bgr = cv2.imread(path)
    return cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)

file_path = 'dog-running.png'
```

```python

def process_image(img_rgb, x1, y1, x2, y2):
    """Process the image and display the results."""
    # img_grayscale = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2GRAY)
    img_array = np.array(img_rgb) / 255.0
    print(img_array.shape)
    img_darker = img_array * 0.5
    img_brighter = img_array * 1.5
    # img_gray = cv2.GaussianBlur(img_gray, (5, 5), 0)

    ROWS = 2
    COLUMNS = 2
    plt.figure(figsize=(10,7))
    plt.subplot(ROWS, COLUMNS, 1)
    plt.axis('off')
    plt.imshow(img_rgb)

    plt.subplot(ROWS, COLUMNS, 2)
    plt.axis('off')
    plt.imshow(img_darker)

    plt.subplot(ROWS, COLUMNS, 3)
    plt.axis('off')
    plt.imshow(img_brighter)

    color = (255, 0, 0)
    thickness = 5
    image_with_rectangle = cv2.rectangle(img_rgb, (x1, y1), (x2, y2), color, thickness)

    plt.subplot(ROWS, COLUMNS, 4)
    plt.axis('off')
    plt.imshow(image_with_rectangle)

    # show all images
    plt.show()


img_rgb = read_image_from_file(file_path)
# Draw the rectangle on the image
x1 = 200
y1 = 150
x2 = 300
y2 = 250
process_image(img_rgb, x1, y1, x2, y2)
```

```python
file_path = 'chest-xray1.jpg'
img = read_image_from_file(file_path)

x1 = 100
y1 = 50
x2 = 200
y2 = 150
process_image(img, x1, y1, x2, y2)
```
