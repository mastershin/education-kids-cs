```python
import matplotlib.pyplot as plt
# Dragon Quest with a Reward of $1 million dollar or one penny today, but doubles every day for 30 days only.
# https://colab.research.google.com/drive/1-Uy244LZOEDEV1GRF_If_ET8IPeKG_K_#scrollTo=e2t9x9vz-ED8
#
# MastershinAI.com MIT License
#
# Quest:
# The princess had been kidnapped by a dragon.
#
# The king offers two reward options for saving princess from the dragon.
# The king does not have a lot of money collected from taxes, but, can offer:
#
# 1. $1 million dollar
#
# 2. one penny today, but doubles every day for 30 days only.
#
# Who is willing to take the challenge?

amount = 0.01
for day in range(30):
  amount = amount * 2
  print(day, amount)
```

```python
# To visualize, rather than printing, save to a list and plot
amount_list = []
days = 30
amount = 0.01
for day in range(days):
  amount = amount * 2
  amount_list.append(amount)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 4))
plt.plot(range(days), amount_list, marker='o')
plt.show()
```
